# Gradient Descent in Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/home/senthil/Data/Data/Data_Base/CSV_file/auto.csv")
```

```python
df = df.dropna()
```

```python
x = df[['city-mpg']]
y = df[['price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
lm.predict(x_test)
```
